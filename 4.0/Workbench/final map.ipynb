```python
import matplotlib.pyplot as plt
import utm
```

```python

loc1 = utm.from_latlon(22.548133, 72.850833)
loc2 = utm.from_latlon(22.433067, 72.991698)

map_plot = plt.imread('crop.png')
bounding_box = (loc1[0]%100000,loc2[0]%100000,loc2[1]%100000,loc1[1]%100000)

home = utm.from_latlon(22.507484, 72.950398)
boat = utm.from_latlon(22.486708, 72.929456)
cbot = utm.from_latlon(22.476398, 72.910745)


latitude = [(home[0])%100000,(boat[0])%100000,(cbot[0])%100000]
longitude = [(home[1])%100000,(boat[1])%100000,(cbot[1])%100000]

fig, ax = plt.subplots(figsize=(20,20))
ax.scatter(latitude[0],longitude[0],color='green',s=400)
ax.scatter(latitude[1],longitude[1],color='blue',s=400)
ax.scatter(latitude[2],longitude[2],color='red',s=400)
ax.imshow(map_plot,extent=bounding_box)
```

```python
from PIL import Image

image = Image.open('crop.png')
width,height = image.size
print(width,height)
```

```python
scale_x = width/abs(loc1[0]-loc2[0])
scale_y = height/abs(loc1[1]-loc2[1])
print(scale_x,scale_y)
```

```python
map_plot = plt.imread('crop.png')
bounding_box = (width,0,height,0)

latitude = [(loc1[0]-home[0])*scale_x+width,(loc1[0]-boat[0])*scale_x+width,(loc1[0]-cbot[0])*scale_x+width]
longitude = [(loc2[1]-home[1])*scale_y+height,(loc2[1]-boat[1])*scale_y+height,(loc2[1]-cbot[1])*scale_y+height]

fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(latitude[0],longitude[0],color='green',s=20)
ax.scatter(latitude[1],longitude[1],color='blue',s=20)
ax.scatter(latitude[2],longitude[2],color='red',s=20)
ax.imshow(map_plot,extent=bounding_box,cmap='gray')
```
