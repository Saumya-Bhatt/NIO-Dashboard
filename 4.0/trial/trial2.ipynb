```python
import PIL
from PIL import Image
import json
import matplotlib.pyplot as plt
from IPython.display import display
from statistics import mean
```

```python
locations = json.loads(open('location.json').read())
print(locations)
```

```python
map_plot = plt.imread('raster_map.png')
```

```python
latitude=[]
longitude=[]
for i in locations:
    latitude.append(locations[i]['latitude'][0])
    longitude.append(locations[i]['longitude'][0])
    
print(latitude,longitude)
```

```python
new_lat = [i-0.00125 for i in latitude]
new_long = [i+0.00125 for i in longitude]

print(new_lat,new_long)
```

```python
bbox2 = (73.79,73.81,15.45,15.46)

fig, ax = plt.subplots(figsize=(15,17))
ax.scatter(new_long[0],new_lat[0],color='green',s=400)
ax.scatter(new_long[1],new_lat[1],color='blue',s=200)
ax.scatter(new_long[2],new_lat[2],color='red',s=100)

ax.imshow(map_plot,extent=bbox2)
```

```python
# importing whole module 
from tkinter import *
from tkinter.ttk import *

# importing strftime function to 
# retrieve system's time 
from time import strftime 

# creating tkinter window 
root = Tk() 
root.title('Clock') 

# This function is used to 
# display time on the label 
def time(): 
    string = strftime('%H:%M:%S %p') 
    lbl.config(text = string) 
    lbl.after(1000, time) 

# Styling the label widget so that clock 
# will look more attractive 
lbl = Label(root, font = ('calibri', 40, 'bold'), 
            background = 'purple', 
            foreground = 'white') 

# Placing clock at the centre 
# of the tkinter window 
lbl.pack(anchor = 'center') 
time() 

```
